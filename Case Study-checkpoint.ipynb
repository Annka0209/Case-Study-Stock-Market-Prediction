```python
import pandas as pd
from datetime import datetime
from pandas_datareader import data
```

```python
start_date = datetime(2007, 7, 1)
end_date = datetime(2017, 7, 1)

symbol = "^STOXX50E"

stoxx50e = data.DataReader(symbol, start=start_date, end=end_date, data_source="yahoo")


testing_start = end_date
testing_end = datetime.now()

stoxx50e_test = data.DataReader(symbol, start=testing_start, end=testing_end, data_source="yahoo")
```

```python
stoxx50e_test
```

```python
from sklearn.linear_model import Ridge
```

```python
import numpy as np
```

```python
close_data = stoxx50e["Close"].fillna(stoxx50e.mean())
regression_data = np.array([
    close_data[i:i+30] for i in range(len(close_data)-31)
])
result_data = close_data[31:]
print(result_data.shape, regression_data.shape)


close_testing = stoxx50e_test["Close"].fillna(stoxx50e_test.mean())
indep_test = np.array([
    close_testing[i:i+30] for i in range(len(close_testing)-31)
])
dep_testing = np.array(close_testing[31:])
```

```python
clf = Ridge(alpha=1.0)
clf.fit(regression_data, result_data)
```

```python
pred = clf.predict(indep_test)
```

```python
err = (pred - dep_testing)**2
np.mean(err)
```
